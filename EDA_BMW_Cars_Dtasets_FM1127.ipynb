```python
# 6. Outlier Detection & Treatment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Load dataset
df = pd.read_csv("/content/BMW Cars Dataset Unchanged.csv")
print(df)
```

```python
# Select numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(df)
```

```python
#  1. IQR (Interquartile Range) Method

print(" Outlier Detection using IQR Method:")

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
    print(f"{col}: {len(outliers)} outliers")

    # Optional: Remove outliers
    df = df[(df[col] >= lower_limit) & (df[col] <= upper_limit)]

print("\n Outliers removed using IQR method!")
```

```python
#  2. Z-Score Method

print("\n Outlier Detection using Z-Score:")

z_scores = np.abs(stats.zscore(df[numeric_cols]))
outlier_mask = (z_scores > 3)   # Threshold of 3 (you can adjust)
print("Number of outliers (Z > 3):", np.sum(outlier_mask))

# Optional: Remove outliers using Z-score
df = df[(z_scores < 3).all(axis=1)]
print("Outliers removed using Z-score!")
```

```python
# 3. Boxplots for Visualization

print("\n Boxplots of Numeric Columns (Before/After Treatment):")
for col in numeric_cols:
    plt.figure(figsize=(5, 3))
    plt.boxplot(df[col], vert=False)
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

```python

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("/content/BMW Cars Dataset Unchanged.csv")

# Select numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# --- Cap Outliers using IQR ---
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower,
                       np.where(df[col] > upper, upper, df[col]))

print(" Outliers capped (Winsorized) using IQR method!")
print(df.shape)
df.head()
```

```python
# 8. Multivariate Analysis (Optional but Recommended)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# --- LOAD DATA ---
df = pd.read_csv("/content/BMW Cars Dataset Unchanged.csv")

# --- CLEAN BASIC ERRORS ---
df = df.dropna()  # drop missing rows for simplicity

# --- IDENTIFY COLUMNS ---
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

print("Numeric Columns:", numeric_cols)
print("Categorical Columns:", categorical_cols)
```

```python
# 1. Pairplot with Hue

# (Run only if you have at least one categorical column)
if len(categorical_cols) > 0 and len(numeric_cols) > 1:
    hue_col = categorical_cols[0]   # use first categorical column (e.g. FuelType)
    sns.pairplot(df, vars=numeric_cols, hue=hue_col, diag_kind='kde', palette='husl')
    plt.suptitle(f"Pairplot with Hue: {hue_col}", y=1.02)
    plt.show()
else:
    print(" Not enough numeric/categorical columns for pairplot.")
```

```python
#  2. Stacked Bar Chart (Categorical vs Categorical)
if len(categorical_cols) >= 2:
    cat1, cat2 = categorical_cols[0], categorical_cols[1]
    cross_tab = pd.crosstab(df[cat1], df[cat2])
    cross_tab.plot(kind='bar', stacked=True, figsize=(7, 5), colormap='Set2')
    plt.title(f"Stacked Bar Chart: {cat1} vs {cat2}")
    plt.ylabel("Count")
    plt.xlabel(cat1)
    plt.legend(title=cat2)
    plt.show()
else:
    print(" Need at least two categorical columns for stacked bar chart.")
```

```python
#  3. PCA (Principal Component Analysis)

if len(numeric_cols) > 2:
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df[numeric_cols])

    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(scaled_data)
    df_pca = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])

    plt.figure(figsize=(7, 5))
    if len(categorical_cols) > 0:
        hue_col = categorical_cols[0]
        sns.scatterplot(x='PC1', y='PC2', data=df_pca, hue=df[hue_col], palette='husl')
    else:
        sns.scatterplot(x='PC1', y='PC2', data=df_pca, color='teal')

    plt.title("PCA Scatter Plot (2 Components)")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.show()

    print("Explained Variance Ratio:", pca.explained_variance_ratio_)
else:
    print(" Not enough numeric columns for PCA.")
```
